````python
from transformers import pipeline

# Load the chat model (same model we used in class)
pipe = pipeline('text-generation', model='TinyLlama/TinyLlama-1.1B-Chat-v1.0')

# Helper function (from class)
def get_response(messages, temperature=0.7, max_new_tokens=100):
    prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    outputs = pipe(prompt, max_new_tokens=max_new_tokens, do_sample=True, temperature=temperature)
    return outputs[0]["generated_text"].split("<|assistant|>")[-1]

print("Setup complete! ✅")
````

### Exercise 1.2: Rules of Engagement — Constraints

System prompts are not just for personas — they are for **rules** too!

**Task:** Write a system prompt that forces the AI to:
1. Answer in **a maximum of 2 sentences** (Constraint)
2. Avoid using **any complex/technical words** (Constraint)

Question: *"How does the internet work?"*


````python
# TODO: Write a system prompt with constraints
messages_rules = [
    {"role": "system", "content": "CRITICAL RULES: You must answer in exactly two sentences or less. Do not use complex, technical, or scientific words under any circumstance. Explain using everyday child-friendly language."},
    {"role": "user", "content": "How does the internet work?"}
]

print(get_response(messages_rules))
````

**🤔 Reflection Question 1.2:** Did the model actually follow your rules? If not, how could you improve the prompt? (Hint: small models often ignore rules — try making your prompt stricter and clearer!)

> *Your answer here...*
Smaller models like TinyLlama-1.1B frequently struggle with strict structural constraints (like sentence counts) because they process text sequentially and lack the capacity to count dynamically. To improve compliance, I can use highly descriptive formatting commands instead of numerical limits (e.g., "Give a very short one-line explanation") and utilize capitalized negative constraints like "NEVER use technical jargon."

## Part 2: Temperature — The "Confidence" Slider 🌡️

Remember:

````python
# Load GPT-2 (for creative text experiments)
generator = pipeline('text-generation', model='gpt2')
print("GPT-2 ready! ✅")
````

### Exercise 2.2: The Repeat Test — Deterministic or Random? 🎲

**Task:** Run the same prompt **3 times at the same temperature**.
- First, 3 runs at `temperature = 0.1`
- Then, 3 runs at `temperature = 1.2`

Observe: in which case are the outputs more similar to each other?


````python
prompt = "The most important invention in history is"

# TODO: loop 3 times at temperature 0.1
for i in range(3):
    res = generator(prompt, max_length=30, do_sample=True, temperature=0.1)
    print(f"Run {i+1}:", res[0]['generated_text'])
    print("-" * 60)
````

````python
# TODO: now 3 times at temperature 1.2
for i in range(3):
    res = generator(prompt, max_length=30, do_sample=True, temperature=1.2)
    print(f"Run {i+1}:", res[0]['generated_text'])
    print("-" * 60)
````

**🤔 Reflection Question 2.2:** Why were the outputs similar at low temperature? Explain in terms of the probability distribution (Hint: remember the "Serious Student" analogy — it only picks the highest-probability word).

> *Your answer here...*
Lowering the temperature sharpens the model's word probability distribution by dramatically amplifying the score of the top tokens while crushing the likelihood of lower-ranked words to near zero. Because the pool is heavily biased toward the single highest-probability choice at each step, the sampling process becomes deterministic, leading to near-identical outputs across separate runs.

## Part 3: Top-P (Nucleus Sampling) — The "Selection Pool" 🎯

Remember:
- **Top-P = 0.1** → the AI only looks at the top 1-2 words (probability sum = 10%)
- **Top-P = 0.9** → a bigger pool, more creativity, but higher chance of mistakes


### Exercise 3.1: Top-P Comparison

**Task:** Same prompt, different top_p values: `0.1`, `0.5`, `0.95`

Prompt: `"In a distant galaxy, humans discovered a planet made of"`


````python
prompt = "In a distant galaxy, humans discovered a planet made of"

# TODO: top_p = 0.1
res = generator(prompt, max_length=50, do_sample=True, top_p=0.1)
print("TOP-P 0.1:\n", res[0]['generated_text'])
````

````python
# TODO: top_p = 0.5
res = generator(prompt, max_length=50, do_sample=True, top_p=0.5)
print("TOP-P 0.5:\n", res[0]['generated_text'])
````

````python
# TODO: top_p = 0.95
res = generator(prompt, max_length=50, do_sample=True, top_p=0.95)
print("TOP-P 0.95:\n", res[0]['generated_text'])
````

**🤔 Reflection Question 3.1:** Temperature and Top-P — both control randomness, but the **mechanism is different**. Explain the difference between the two in your own words:

> *Your answer here...*
Temperature controls the shape of the entire token probability distribution curve (flattening it out to add chaos or sharpening it to add focus). Top-P (Nucleus Sampling) controls the cutoff boundary of the token pool, dynamically truncating the long tail of low-probability words once the cumulative percentage threshold is hit. Temperature adjusts the relative weights of words, whereas Top-P physically drops words from consideration.
